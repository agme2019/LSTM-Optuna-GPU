```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
from tensorflow import keras
import tensorflow as tf
import optuna
from optuna.samplers import TPESampler
from optuna.pruners import MedianPruner
import time
import io

# Check and configure GPU
print("GPU Available:", tf.config.list_physical_devices('GPU'))
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    try:
        # Configure for growth to prevent memory fragmentation
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        # Enable mixed precision for faster computation on A100
        policy = keras.mixed_precision.Policy('mixed_float16')
        keras.mixed_precision.set_global_policy(policy)
        print("Mixed precision enabled")
    except RuntimeError as e:
        print(e)

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Start timing the execution
start_time = time.time()

# 1. Load and Prepare the Data
data = pd.read_csv('AAPL.csv')
print(data.head())
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)
data.sort_index(inplace=True)

# Extract relevant columns
features = ['Open', 'High', 'Low', 'Close', 'Volume']
data = data[features]

# 2. Scale the Data - Feature-specific scaling
scaled_data = np.zeros_like(data.values)
scalers = {}

for i, feature in enumerate(features):
    scalers[feature] = MinMaxScaler(feature_range=(0, 1))
    feature_data = data[feature].values.reshape(-1, 1)
    scaled_data[:, i] = scalers[feature].fit_transform(feature_data).flatten()

print("Data scaled using feature-specific scalers")

# 3. Create Sequences with TensorFlow operations for speed
def create_dataset(dataset, time_step=60):
    X, y = [], []
    for i in range(len(dataset) - time_step):
        X.append(dataset[i:i+time_step])
        y.append(dataset[i+time_step, 3])  # Close price
    return np.array(X), np.array(y)

time_step = 60
X_all, y_all = create_dataset(scaled_data, time_step=time_step)
all_dates = data.index[time_step:]

# 4. Time-based Train-Validation-Test Split
train_size = int(len(X_all) * 0.7)
val_size = int(len(X_all) * 0.15)

X_train, y_train = X_all[:train_size], y_all[:train_size]
X_val, y_val = X_all[train_size:train_size+val_size], y_all[train_size:train_size+val_size]
X_test, y_test = X_all[train_size+val_size:], y_all[train_size+val_size:]

train_dates = all_dates[:train_size]
val_dates = all_dates[train_size:train_size+val_size]
test_dates = all_dates[train_size+val_size:]

print(f"Training: {X_train.shape[0]} samples ({train_dates.min()} to {train_dates.max()})")
print(f"Validation: {X_val.shape[0]} samples ({val_dates.min()} to {val_dates.max()})")
print(f"Test: {X_test.shape[0]} samples ({test_dates.min()} to {test_dates.max()})")

# 5. Create TF Datasets for more efficient training
batch_size = 256  # Large batch size for A100

train_dataset = tf.data.Dataset.from_tensor_slices((X_train, y_train))
train_dataset = train_dataset.shuffle(buffer_size=1024).batch(batch_size).prefetch(tf.data.AUTOTUNE)

val_dataset = tf.data.Dataset.from_tensor_slices((X_val, y_val))
val_dataset = val_dataset.batch(batch_size).prefetch(tf.data.AUTOTUNE)

test_dataset = tf.data.Dataset.from_tensor_slices((X_test, y_test))
test_dataset = test_dataset.batch(batch_size).prefetch(tf.data.AUTOTUNE)

# 6. Define model creation function with expanded hyperparameter space
def create_model(trial):
    # Create model using Keras Functional API
    inputs = keras.layers.Input(shape=(X_train.shape[1], X_train.shape[2]))

    # First LSTM layer - expanded parameter space
    units_1 = trial.suggest_categorical('units_1', [64, 128, 192, 256, 320])
    activation_1 = trial.suggest_categorical('activation_1', ['tanh', 'relu', 'selu'])
    dropout_1 = trial.suggest_float('dropout_1', 0.1, 0.6)

    x = keras.layers.LSTM(
        units=units_1,
        activation=activation_1,
        return_sequences=True,
        recurrent_dropout=0.0  # Better GPU performance
    )(inputs)
    x = keras.layers.Dropout(dropout_1)(x)

    # Possible attention mechanism
    use_attention = trial.suggest_categorical('use_attention', [True, False])
    if use_attention:
        # Simple attention mechanism
        attention = keras.layers.Dense(1)(x)
        attention = keras.layers.Reshape((X_train.shape[1],))(attention)
        attention = keras.layers.Activation('softmax')(attention)
        attention = keras.layers.RepeatVector(units_1)(attention)
        attention = keras.layers.Permute([2, 1])(attention)
        x = keras.layers.Multiply()([x, attention])

    # Second LSTM layer
    units_2 = trial.suggest_categorical('units_2', [64, 128, 192, 256])
    activation_2 = trial.suggest_categorical('activation_2', ['tanh', 'relu', 'selu'])
    dropout_2 = trial.suggest_float('dropout_2', 0.1, 0.6)
    return_sequences = trial.suggest_categorical('return_sequences', [True, False])

    x = keras.layers.LSTM(
        units=units_2,
        activation=activation_2,
        return_sequences=return_sequences,
        recurrent_dropout=0.0
    )(x)
    x = keras.layers.Dropout(dropout_2)(x)

    # Optional third LSTM layer
    use_third_layer = trial.suggest_categorical('use_third_layer', [True, False])
    if use_third_layer and return_sequences:
        units_3 = trial.suggest_categorical('units_3', [32, 64, 96, 128])
        activation_3 = trial.suggest_categorical('activation_3', ['tanh', 'relu', 'selu'])
        x = keras.layers.LSTM(
            units=units_3,
            activation=activation_3,
            recurrent_dropout=0.0
        )(x)
        x = keras.layers.Dropout(trial.suggest_float('dropout_3', 0.1, 0.5))(x)

    # If we still have sequences at this point, flatten them
    if return_sequences and not (use_third_layer and return_sequences):
        x = keras.layers.Flatten()(x)

    # Dense layers configuration
    dense_layers = trial.suggest_int('dense_layers', 0, 2)
    for i in range(dense_layers):
        units = trial.suggest_categorical(f'dense_units_{i}', [16, 32, 64, 128])
        x = keras.layers.Dense(units, activation='relu')(x)
        x = keras.layers.Dropout(trial.suggest_float(f'dense_dropout_{i}', 0.1, 0.4))(x)

    # Output layer
    outputs = keras.layers.Dense(1)(x)

    # Create and compile model
    model = keras.Model(inputs=inputs, outputs=outputs)

    # Optimizer selection - expanded options
    optimizer_name = trial.suggest_categorical('optimizer', ['adam', 'rmsprop', 'adamw'])
    learning_rate = trial.suggest_float('learning_rate', 5e-5, 5e-2, log=True)

    if optimizer_name == 'adam':
        optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
    elif optimizer_name == 'rmsprop':
        optimizer = keras.optimizers.RMSprop(learning_rate=learning_rate)
    else:  # adamw
        optimizer = keras.optimizers.AdamW(learning_rate=learning_rate, weight_decay=trial.suggest_float('weight_decay', 1e-6, 1e-3, log=True))

    # Loss function selection
    loss_func = trial.suggest_categorical('loss', ['mean_squared_error', 'mean_absolute_error', 'huber'])
    model.compile(optimizer=optimizer, loss=loss_func)
    return model

# 7. Objective function for hyperparameter optimization with more options
def objective(trial):
    # Create model with trial hyperparameters
    model = create_model(trial)

    # Early stopping callback with dynamic patience
    patience = trial.suggest_int('patience', 3, 10)
    early_stopping = keras.callbacks.EarlyStopping(
        monitor='val_loss',
        patience=patience,
        restore_best_weights=True
    )

    # Dynamic epochs and batch size
    epochs = trial.suggest_int('epochs', 15, 50)

    # Train the model
    history = model.fit(
        train_dataset,
        epochs=epochs,
        validation_data=val_dataset,
        callbacks=[early_stopping],
        verbose=0
    )

    # Return best validation loss
    return min(history.history['val_loss'])

# 8. Run Optuna optimization with more trials
print("Starting hyperparameter optimization with extended trials...")
study = optuna.create_study(
    direction='minimize',
    sampler=TPESampler(n_startup_trials=15),  # More startup trials for better exploration
    pruner=MedianPruner(n_warmup_steps=5)
)

# Increase to 50 trials for more thorough exploration
n_trials = 50
print(f"Running {n_trials} trials for hyperparameter optimization...")
study.optimize(objective, n_trials=n_trials, timeout=7200)  # 2 hour timeout

best_params = study.best_params
best_value = study.best_value
print(f"Best trial value: {best_value}")
print("Best Hyperparameters:", best_params)

# 9. Advanced analysis of hyperparameter importance
print("\nHyperparameter Importance:")
param_importance = optuna.importance.get_param_importances(study)
for param, importance in param_importance.items():
    print(f"{param}: {importance:.4f}")

# 10. Train the final model with best parameters
print("\nTraining final model with best parameters...")
final_model = create_model(optuna.trial.FixedTrial(best_params))

# Callbacks for the final model
early_stopping = keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=best_params.get('patience', 8),
    restore_best_weights=True
)

# Model checkpoint to save the best model
model_checkpoint = keras.callbacks.ModelCheckpoint(
    'best_model_a100_extended.h5',
    monitor='val_loss',
    save_best_only=True,
    mode='min'
)

# Reduce LR on plateau for fine-tuning
reduce_lr = keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=3,
    min_lr=1e-6,
    verbose=1
)

# Training with TensorBoard for visualization
tensorboard = keras.callbacks.TensorBoard(
    log_dir='./logs',
    histogram_freq=1
)

# Train the final model
history = final_model.fit(
    train_dataset,
    epochs=best_params.get('epochs', 30) + 10,  # Train a bit longer for final model
    validation_data=val_dataset,
    callbacks=[early_stopping, model_checkpoint, reduce_lr, tensorboard],
    verbose=1
)

# 11. Evaluate on test set
print("Evaluating on test set...")
y_pred = final_model.predict(test_dataset)

# Inverse transform predictions and actual values
y_pred_inv = scalers['Close'].inverse_transform(y_pred)
y_test_array = np.array(list(test_dataset.unbatch().map(lambda x, y: y).as_numpy_iterator()))
y_test_inv = scalers['Close'].inverse_transform(y_test_array.reshape(-1, 1))

# Calculate metrics
mse = mean_squared_error(y_test_inv, y_pred_inv)
rmse = np.sqrt(mse)
mape = mean_absolute_percentage_error(y_test_inv, y_pred_inv) * 100

print(f"Mean Squared Error: {mse:.4f}")
print(f"Root Mean Squared Error: {rmse:.4f}")
print(f"Mean Absolute Percentage Error: {mape:.2f}%")

# 12. Visualize results with improved plots
plt.figure(figsize=(14, 7))
plt.plot(test_dates, y_test_inv, 'b-', label='Actual Price', linewidth=2, alpha=0.7)
plt.plot(test_dates, y_pred_inv, 'r-', label='Predicted Price', linewidth=2, alpha=0.7)
plt.title('Stock Price Prediction (Extended Trials A100 Model)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('prediction_a100_extended.png', dpi=300)

# 13. More detailed error analysis
residuals = y_test_inv - y_pred_inv

plt.figure(figsize=(16, 12))
# Residuals over time
plt.subplot(2, 2, 1)
plt.plot(test_dates, residuals, 'g-', linewidth=1.5)
plt.title('Prediction Residuals Over Time', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Error', fontsize=12)
plt.grid(True, alpha=0.3)

# Histogram of residuals
plt.subplot(2, 2, 2)
plt.hist(residuals, bins=50, color='purple', alpha=0.7)
plt.title('Distribution of Prediction Errors', fontsize=14)
plt.xlabel('Error', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True, alpha=0.3)

# Predicted vs Actual scatter plot
plt.subplot(2, 2, 3)
plt.scatter(y_test_inv, y_pred_inv, alpha=0.5, color='blue')
plt.plot([y_test_inv.min(), y_test_inv.max()], [y_test_inv.min(), y_test_inv.max()], 'r--')
plt.title('Predicted vs Actual Prices', fontsize=14)
plt.xlabel('Actual Price', fontsize=12)
plt.ylabel('Predicted Price', fontsize=12)
plt.grid(True, alpha=0.3)

# Percentage Error plot
pct_errors = (residuals / y_test_inv) * 100
plt.subplot(2, 2, 4)
plt.plot(test_dates, pct_errors, color='orange', linewidth=1.5)
plt.title('Percentage Error Over Time', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Percentage Error (%)', fontsize=12)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('error_analysis_a100_extended.png', dpi=300)

# 14. Plot training history with more details
plt.figure(figsize=(14, 8))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss', linewidth=2)
plt.plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
plt.title('Training and Validation Loss', fontsize=14)
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Loss', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)

# Plot learning rate if available
if 'lr' in history.history:
    plt.subplot(1, 2, 2)
    plt.plot(history.history['lr'], label='Learning Rate', linewidth=2, color='green')
    plt.title('Learning Rate Schedule', fontsize=14)
    plt.xlabel('Epochs', fontsize=12)
    plt.ylabel('Learning Rate', fontsize=12)
    plt.yscale('log')
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('training_history_a100_extended.png', dpi=300)

# Print execution time
execution_time = time.time() - start_time
print(f"Total execution time: {execution_time:.2f} seconds ({execution_time/60:.2f} minutes)")

# Save summary statistics and hyperparameter details to file
# FIX: Properly handle file opening and closing
with open('model_summary_a100_extended.txt', 'w') as f:
    f.write(f"Model Summary for A100-Optimized LSTM with Extended Trials\n")
    f.write(f"======================================================\n")
    f.write(f"Number of trials: {n_trials}\n")
    f.write(f"Best Hyperparameters:\n")
    for param, value in best_params.items():
        f.write(f"  {param}: {value}\n")
    f.write(f"\nPerformance Metrics:\n")
    f.write(f"  Mean Squared Error: {mse:.4f}\n")
    f.write(f"  Root Mean Squared Error: {rmse:.4f}\n")
    f.write(f"  Mean Absolute Percentage Error: {mape:.2f}%\n")
    f.write(f"  Total Execution Time: {execution_time/60:.2f} minutes\n")

    f.write(f"\nHyperparameter Importance:\n")
    for param, importance in param_importance.items():
        f.write(f"  {param}: {importance:.4f}\n")

    # Capture model summary separately
    summary_string = io.StringIO()
    final_model.summary(print_fn=lambda x: summary_string.write(x + '\n'))
    f.write(f"\nModel Architecture:\n")
    f.write(summary_string.getvalue())

# 15. Save trial history for future analysis
trials_df = study.trials_dataframe()
trials_df.to_csv('optuna_trials_history.csv', index=False)

# Generate a parallel coordinate plot for visualization
try:
    from optuna.visualization import plot_parallel_coordinate
    import plotly.io as pio

    fig = plot_parallel_coordinate(study)
    pio.write_image(fig, 'hyperparameter_parallel_coordinates.png')
    print("Saved hyperparameter parallel coordinates plot")
except Exception as e:
    print(f"Could not create parallel coordinate plot: {e}")
```
